```python
#melakukan importing modul pada python untuk memroses dataset
import pandas 
import numpy as np
import sklearn
import operator
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#melakukan pembacaan dataset yang akan diolah dan dibuat model
#dataset disimpan pada variabel df
df = pandas.read_csv("el4233-2018-2019-02-01-training.csv", header=None)
print(df)
```

```python
#melakukan plotting data dasar dari dataset dengan menggunakan model scatter plot
plt.scatter(df[0], df[1])
plt.show()
```

```python
#melakukan splitting data dasar menjadi data training dan data test dengan porsi 80:20
#setelah didapat data training maka dibuat ke 2D array
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(df[0], df[1], test_size=0.20)

xtrain = xtrain[:,np.newaxis]
xtest = xtest[:,np.newaxis]
xtrain.shape
xtest.shape
ytrain.shape
```

```python
#melakukan plotting data dari data training dan data test
plt.plot(xtrain, ytrain, "bo",label='training set') # training set
plt.plot(xtest, ytest, "gx",label='validation set') # validation set

plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, -2, 3])
plt.show()
```

```python
#menginisiasi dua buah array (train dan test) ukuran 100 yang isinya 0
report_train_mse=np.zeros(100)
report_test_mse=np.zeros(100)
```

```python
#melakukan pembuatan model polinomial dengan menggunakan modul sklearn dengan derajat 15
derajat_polinom=15
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=derajat_polinom, include_bias=False)
xtrain_poly = poly_features.fit_transform(xtrain)

#model yang didapat dibuatkan ke model regresi linear dengan variabel y adalah ytrain
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(xtrain_poly, ytrain)
    
ytrain_predict=lin_reg.predict(xtrain_poly)
```

```python
#membaca intercept dan koefisien dari model regresi linear
lin_reg.intercept_, lin_reg.coef_
```

```python
#membuat plot dari grafik prediksi model dari test set
plt.scatter(xtrain, ytrain_predict, color='k', label="prediction")
plt.plot(xtest, ytest, "gx",label='validation set') # test set
plt.plot(xtrain, ytrain, "bo",label='training set') # training set

plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, -2,3])

plt.legend(loc='lower left')
plt.show()
```

```python
#menghitung mse dari training set dan test set
train_mse=sklearn.metrics.mean_squared_error(ytrain, ytrain_predict)
xtest_poly = poly_features.fit_transform(xtest)
ytest_predict=lin_reg.predict(xtest_poly)
test_mse=sklearn.metrics.mean_squared_error(ytest, ytest_predict)
```

```python
#memasukan mse dari training set ke array
report_train_mse[derajat_polinom]=train_mse
report_test_mse[derajat_polinom]=test_mse
report_train_mse
```

```python
#memasukan mse dari test set ke array
report_test_mse
```
