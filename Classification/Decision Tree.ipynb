```python
#melakukan import module yang diperlukan untuk melakukan decision tree algorithm
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier # import Decision Tree Classifier
from sklearn.model_selection import train_test_split # import fungsi train_test_split
from sklearn import metrics #import modul metrics dari scikit-learn untuk mengkalkulasi akurasi
```

```python
#membaca dataset yang akan dilakukan pemodelan
df = pd.read_csv("el4233-2018-2019-02-klasifikasi-train.csv")
df1 = df.iloc[:, 1:4] #tidak membaca bagian index yang berasal dari dataset asli
```

```python
#melakukan plotting terhadap feature yang ada beserta target hasilnya
X = df1.iloc[:, 0:2] #assign X sebagai X0 dan X1, feature dari model 
y = df1.iloc[:, 2] #assign Y sebagai Y, output dari feature yang ada

yes = df1.loc[y==1] #klasifikasikan sebagai 'yes' ketika y==1
no = df1.loc[y ==0] #klasifikasikan sebagai 'no' ketika y==0

#plotting dengan scatter
plt.scatter(yes.iloc[:, 0], yes.iloc[:, 1], s=10, label='yes')
plt.scatter(no.iloc[:, 0], no.iloc[:, 1], s=10, label='no')
plt.legend()
plt.show()
```

```python
#melakukan splitting data asli menjadi train dan test dengan rasio 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#membuat objek dari Decision Tree Classsifier
clf = DecisionTreeClassifier()

# train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#memprediksi jawaban dari dataset test
y_pred = clf.predict(X_test)
```

```python
#mengecek akurasi dari model 
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
